```python
#no easy way to distinguish trains from busses, plus some trainlines are temporaily served by buw

dfSFLines = json_normalize(apsl)
SFTrains = []
for i in range (len(dfSFLines.index)):
    if dfSFLines.loc[i, 'TransportMode'] == 'metro':
        SFTrains.append(i)
        
    if  dfSFLines.loc[i, "PublicCode"].find('BUS') != -1:
    
        SFTrains.append(i)
        
dfSFTrains = dfSFLines.loc[SFTrains]
dfSFTrains.reset_index(inplace= True, drop = True)
stopLib = []
for i in range(len(dfSFTrains)):
        url511 = url511pd.format('stops', key511, '&operator_id=SF&line_id={}'.format(dfSFTrains.loc[i,'PublicCode']))
        response = requests.get(url511).text.encode().decode('utf-8-sig')
        responseC = json.loads(response)
        stopLib.append(responseC)

        
                                 
```

```python
dfClose = dfRests
dfClose['Closest Station']= ''
dfClose['Distance from Station'] = 1000
dfClose['slat'] = 0
dfClose['slon'] = 0
```

```python
for i in range(len(dfClose.index)):
    for j in range(len(dfStops.index)):
        d1 = [dfClose.loc[ i, 'location.lat'], dfClose.loc[i, 'location.lng']]
        d2 = [dfStops.loc[j, 'Location.Latitude'], dfStops.loc[j, 'Location.Longitude']]
        dis = geodesic(d2, d1).mi
        if dis < dfClose.loc[i,'Distance from Station']:
            dfClose.loc[i,'Distance from Station'] = dis
            dfClose.loc[i , 'Closest Station'] = dfStops.loc[j, 'Name']
            dfClose.loc[i, 'slat'] = dfStops.loc[j, 'Location.Latitude']
            dfClose.loc[i,'slon'] = dfStops.loc[j, 'Location.Longitude']
dfClose

tooFar = []
for i in range (len(dfClose.index)):
    if dfClose.loc[i, 'Distance from Station'] > 5:
        tooFar.append(i)
```

```python

def getDir(dfDir, place):
    
    libDir = []
            
    for i in range(len(dfDir)):

        
        toTrain = 10000
        iClo = 100000
        count = 0
        for j in range(len(dfDir)):

            if dfDir.loc[j, 'Distance from Station'] < toTrain:
                iClo = j
                toTrain = dfDir.loc[j, 'Distance from Station']

        b, c = dfDir.iloc[0].copy(), dfDir.iloc[iClo].copy()
        dfDir.iloc[0],dfDir.iloc[iClo] = c,b

        brkt = '{},{}'
        if i == 0:
            fromll = brkt.format(dfDir.loc[0, 'slat'], dfDir.loc[0, 'slon'])
            start = 'Station'
        else:
            fromll = brkt.format(dfDir.loc[i-1, 'location.lat'], dfDir.loc[i-1, 'location.lng'])
            start = dfDir.loc[i-1, 'name']

        toll = brkt.format(dfDir.loc[i,'location.lat'], dfDir.loc[i ,'location.lng'])
        end = dfDir.loc[i, 'name']


        urlMpd = 'https://open.mapquestapi.com/directions/v2/route?key={}&from={}&to={}'
        urlM = urlMpd.format(keymapq, fromll, toll)
        walk = requests.get(urlM).json()

        w= walk['route']['legs'][0]['maneuvers']
        dfwalk = json_normalize(w)
        
        keyspd = 'leg {} from {} to {}'
        keytitle = keyspd.format(i+1, start, end)
        libDir.append([keytitle, dfwalk])
    libFinal[place] = libDir

```
